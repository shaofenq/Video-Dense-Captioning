```python
compare_res = {}
for key in res.keys():
  if key in gt_data.keys():
      dict_t = { "gt_cap": gt_data[key], "our_cap":res[key] , "baseline_cap":res_baseline[key]}
      compare_res[key] = dict_t
```

```python
len(compare_res)
```

```python
compare_res['v_kBh_98QimD4']
```

```python
compare_res['v_Aygp8JaMkqQ']
```

```python
list(compare_res.keys())[3962]
```

```python
sentence_scores = []
for key in res.keys():
  if key in compare_res.keys():
    curr_video_cap = res[key]
    for cap in curr_video_cap:
      curr_cap_dur = cap['sentence_score']
      sentence_scores.append(curr_cap_dur)
```

```python
sentence_scores_base = []
for key in res.keys():
  if key in compare_res.keys():
    curr_video_cap = res_baseline[key]
    for cap in curr_video_cap:
      curr_cap_dur = cap['sentence_score']
      sentence_scores_base.append(curr_cap_dur)
```

```python
import matplotlib.pyplot as plt
plt.hist(sentence_scores)
plt.title('Sentence Score Distribution for Our Multi-head Attention Model')
plt.ylabel('Sentence #')
plt.xlabel('Sentence Score')
plt.show()
```

```python
plt.hist(sentence_scores_base)
plt.title('Sentence Score Distribution for Baseline model')
plt.ylabel('Sentence #')
plt.xlabel('Sentence Score')
plt.show()
```
